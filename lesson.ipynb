## Outline:

### Section 1: Introduction to Linear Regression

#### Definition
Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables. It assumes a linear relationship between the variables, where changes in the independent variable(s) result in a linear change in the dependent variable.

#### Applications

Application in the Notebook: Predicting Fuel Efficiency (MPG) of Automobiles

In this notebook, we applied linear regression to predict the fuel efficiency (miles per gallon, MPG) of automobiles based on various features such as weight, engine displacement, and horsepower.Linear regression was used to model the relationship between these features and the dependent variable (MPG), allowing us to make predictions about a vehicle's fuel efficiency

### Section 2: 
#### Dependent and Independent Variables

Dependent Variable (Response Variable): This is the variable we are trying to predict or explain. In the context of our Auto MPG example, the dependent variable is the fuel efficiency of automobiles, typically measured in miles per gallon (MPG). It's the outcome or result we want to understand or forecast.

Independent Variable (Predictor Variable): These are the variables that are used to make predictions or explain the variation in the dependent variable. In our case, independent variables could include features like the weight of the vehicle, engine displacement, or horsepower. These variables are assumed to influence or have a relationship with the dependent variable.

#### Linear Relationship

In linear regression we assume that the relationship between the dependent and independent variables is linear. This means that the change in the dependent variable is directly proportional to changes in the independent variable, resulting in a straight-line relationship when plotting on a graph.

For example, in the Auto MPG dataset, we assume that an increase in vehicle weight is associated with a linear change in fuel efficiency. If this assumption holds, we can represent this relationship as a straight line using the equation y = mx + b, where y is the dependent variable (MPG), x is the independent variable (weight), m is the slope of the line, and b is the y-intercept

### Section 3: Mathematical Representation

#### Equation of a Line
- y = mx + b, where y is the dependent variable, x is the independent variable, m is the slope, and b is the y-intercept.

#### Univariate Linear Regression Equation
- Formulation: y = β₀ + β₁x + ε
- Breakdown of Terms:
  - y: Dependent variable.
  - x: Independent variable.
  - β₀: Intercept term (y-intercept).
  - β₁: Slope coefficient.
  - ε: Error term (residuals).

### Section 4: Parameter Estimation

#### Method of Least Squares
in linear regression, we want to find a straight line that best fits our data. This line is described by the equation: y = β₀ + β₁x.

β₀ (Beta Zero): This is the starting point on the y-axis when x is zero.
β₁ (Beta One): This is the slope of the line, telling us how much y changes for a one-unit change in x.

The objective is to find the values of β₀ and β₁ that make the regression line the best fit for the data. The best fit means that the line Minimize: Σ(yᵢ - ŷᵢ)²
